```python
import requests
from bs4 import BeautifulSoup
import re

def extract_nutrition_from_url(url):
    # Usamos un User-Agent para simular un navegador normal y evitar que la página nos bloquee
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status() # Verifica si hubo errores HTTP
    except requests.exceptions.RequestException as e:
        return f"Error al acceder a la página: {e}"

    soup = BeautifulSoup(response.text, 'html.parser')

    # Buscamos todos los elementos <p> que en su clase contengan 'recipe-nutrition__item'
    # Usamos re.compile para ignorar sufijos aleatorios generados por Svelte
    nutrition_items = soup.find_all('p', class_=re.compile(r'recipe-nutrition__item'))

    if not nutrition_items:
        return "No se encontró información nutricional en el formato esperado."

    print(f"--- Información Nutricional extraída ---")

    extracted_data = []
    for item in nutrition_items:
        # Obtenemos el texto, lo limpiamos de espacios y saltos de línea innecesarios
        text = item.get_text(separator=' ', strip=True)
        text = re.sub(r'\s+', ' ', text)
        extracted_data.append(text)
        print(text)

    return extracted_data

# Probando la función con la primera receta que extrajimos antes
print(f"Extrayendo de: {recipe_url}\n")
nutrition_info = extract_nutrition_from_url(recipe_url)
print(nutrition_info)
```

### Extracción Masiva de Serving Size
A continuación, crearemos una función para iterar sobre el dataset y extraer el `servingSize` y `servingPerRecipe` usando el endpoint `/as-json`.

```python
import time
import re
import requests
from tqdm.auto import tqdm # Para una barra de progreso

def get_serving_info_from_url(url, headers):
    """Función auxiliar para intentar extraer el JSON de una URL dada."""
    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            try:
                data = response.json()
                nutrition = data.get('recipe', {}).get('nutrition', {})
                return nutrition.get('servingSize'), nutrition.get('servingPerRecipe')
            except ValueError:
                return None, None # No es un JSON válido
    except Exception:
        return None, None
    return None, None

def extract_serving_info_batch(df, limit=100):
    """
    Extrae la información de porciones para un lote de recetas manejando redirecciones.
    """
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}

    # Crear nuevas columnas si no existen
    if 'ExtractedServingSize' not in df.columns:
        df['ExtractedServingSize'] = None
    if 'ExtractedServings' not in df.columns:
        df['ExtractedServings'] = None

    # Iterar sobre las primeras 'limit' filas
    for index, row in tqdm(df.head(limit).iterrows(), total=limit):
        recipe_id = row['RecipeId']
        recipe_name = row['Name']

        # Reconstruir la URL base inferida
        clean_name = str(recipe_name).lower().replace("'", "")
        slug = re.sub(r'[^a-z0-9]+', '-', clean_name).strip('-')
        base_url = f"https://www.food.com/recipe/{slug}-{recipe_id}"
        json_url = f"{base_url}/as-json"

        # Intento 1: Con la URL inferida
        serving_size, servings = get_serving_info_from_url(json_url, headers)

        # Intento 2: Si falló, la URL pudo haber cambiado. Dejamos que Food.com nos redirija a la URL correcta.
        if serving_size is None and servings is None:
            try:
                # Hacemos petición a la página HTML para obtener la URL final tras la redirección
                res_html = requests.get(base_url, headers=headers, allow_redirects=True, timeout=10)
                canonical_url = res_html.url.rstrip('/')

                # Si hubo redirección a una URL distinta de recetas, intentamos de nuevo
                if canonical_url != base_url and "/recipe/" in canonical_url:
                    canonical_json_url = f"{canonical_url}/as-json"
                    serving_size, servings = get_serving_info_from_url(canonical_json_url, headers)
            except Exception as e:
                pass

        # Actualizar el DataFrame
        df.at[index, 'ExtractedServingSize'] = serving_size
        df.at[index, 'ExtractedServings'] = servings

        # Pausa para ser amables con el servidor
        time.sleep(0.5)

    return df

# Ejecutar la prueba con 50 recetas
print("Extrayendo datos de las primeras 50 recetas (con manejo de redirecciones)...")
recipes = extract_serving_info_batch(recipes, limit=50)

# Mostrar los resultados (Observa la receta 44)
display(recipes[['RecipeId', 'Name', 'ExtractedServingSize', 'ExtractedServings']].head(10))
```

### Extracción Acelerada (Multithreading)
Esta versión utiliza múltiples hilos para hacer peticiones en paralelo. Es mucho más rápido, pero **no abuses de `MAX_WORKERS`** o tu IP será bloqueada por el servidor.

```python
import concurrent.futures
import time
import re
import requests
from tqdm.auto import tqdm
import pandas as pd

# Aprovechando Colab Pro: Lote mucho más grande y un poco más de hilos
TAMANO_LOTE = 50000
MAX_WORKERS = 10

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}

mask_pendientes = recipes['ExtractedServingSize'].isna()
indices_pendientes = recipes[mask_pendientes].head(TAMANO_LOTE).index

print(f"Procesando {len(indices_pendientes)} recetas con {MAX_WORKERS} hilos en paralelo...")

def process_recipe(index):
    row = recipes.loc[index]
    recipe_id = row['RecipeId']
    recipe_name = row['Name']

    clean_name = str(recipe_name).lower().replace("'", "")
    slug = re.sub(r'[^a-z0-9]+', '-', clean_name).strip('-')
    base_url = f"https://www.food.com/recipe/{slug}-{recipe_id}"
    json_url = f"{base_url}/as-json"

    serving_size, servings = get_serving_info_from_url(json_url, headers)

    if serving_size is None and servings is None:
        try:
            res_html = requests.get(base_url, headers=headers, allow_redirects=True, timeout=10)
            canonical_url = res_html.url.rstrip('/')
            if canonical_url != base_url and "/recipe/" in canonical_url:
                canonical_json_url = f"{canonical_url}/as-json"
                serving_size, servings = get_serving_info_from_url(canonical_json_url, headers)
        except Exception:
            pass

    return index, serving_size, servings

# Usar ThreadPoolExecutor para paralelizar
with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
    # Enviar las tareas al pool de hilos
    futures = {executor.submit(process_recipe, idx): idx for idx in indices_pendientes}

    # Recolectar resultados a medida que se completan (y mostrar en la barra de progreso)
    for future in tqdm(concurrent.futures.as_completed(futures), total=len(indices_pendientes)):
        idx, serving_size, servings = future.result()
        recipes.at[idx, 'ExtractedServingSize'] = serving_size
        recipes.at[idx, 'ExtractedServings'] = servings

# Guardar progreso
backup_path = 'recipes_backup.csv'
recipes.to_csv(backup_path, index=False)
print(f"\n¡Lote completado mucho más rápido! Progreso guardado en '{backup_path}'.")
```

```python
import concurrent.futures
import time
import re
import requests
from tqdm.auto import tqdm
import pandas as pd

MAX_WORKERS = 10
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}

# Tomar TODOS los índices que aún faltan por procesar
mask_pendientes = recipes['ExtractedServingSize'].isna()
indices_pendientes = recipes[mask_pendientes].index

print(f"Procesando TODAS las {len(indices_pendientes)} recetas restantes con {MAX_WORKERS} hilos en paralelo...")

def process_recipe(index):
    row = recipes.loc[index]
    recipe_id = row['RecipeId']
    recipe_name = row['Name']

    clean_name = str(recipe_name).lower().replace("'", "")
    slug = re.sub(r'[^a-z0-9]+', '-', clean_name).strip('-')
    base_url = f"https://www.food.com/recipe/{slug}-{recipe_id}"
    json_url = f"{base_url}/as-json"

    serving_size, servings = get_serving_info_from_url(json_url, headers)

    if serving_size is None and servings is None:
        try:
            res_html = requests.get(base_url, headers=headers, allow_redirects=True, timeout=10)
            canonical_url = res_html.url.rstrip('/')
            if canonical_url != base_url and "/recipe/" in canonical_url:
                canonical_json_url = f"{canonical_url}/as-json"
                serving_size, servings = get_serving_info_from_url(canonical_json_url, headers)
        except Exception:
            pass

    return index, serving_size, servings

# Guardado periódico para no perder el progreso si se interrumpe
save_interval = 50000
processed_count = 0

with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
    futures = {executor.submit(process_recipe, idx): idx for idx in indices_pendientes}

    for future in tqdm(concurrent.futures.as_completed(futures), total=len(indices_pendientes)):
        idx, serving_size, servings = future.result()
        recipes.at[idx, 'ExtractedServingSize'] = serving_size
        recipes.at[idx, 'ExtractedServings'] = servings

        processed_count += 1
        if processed_count % save_interval == 0:
            recipes.to_csv('recipes_backup_in_progress.csv', index=False)
            print(f"Guardado intermedio: {processed_count} recetas procesadas.")

# Guardar progreso final
backup_path = 'recipes_backup_final.csv'
recipes.to_csv(backup_path, index=False)
print(f"\n¡Extracción total completada! Dataset completo guardado en '{backup_path}'.")
```

## Resolucion recetas que no contienen ExtractedServingSize, despues del proceso de extraccion

```python
import pandas as pd
import requests
import re
import time
import concurrent.futures
from google.colab import files
from tqdm.auto import tqdm

# 1. Identificar las recetas que siguen fallando o no fueron encontradas
mask_pendientes = recipes['ExtractedServingSize'].isna() | (recipes['ExtractedServingSize'] == "NOT_FOUND")
indices_pendientes = recipes[mask_pendientes].index

print(f"Intentando raspar de nuevo las {len(indices_pendientes)} recetas que fallaron con 10 hilos...")
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
MAX_WORKERS = 10

def process_failed_recipe(index):
    row = recipes.loc[index]
    recipe_id = row['RecipeId']
    recipe_name = row['Name']

    clean_name = str(recipe_name).lower().replace("'", "")
    slug = re.sub(r'[^a-z0-9]+', '-', clean_name).strip('-')
    base_url = f"https://www.food.com/recipe/{slug}-{recipe_id}"
    json_url = f"{base_url}/as-json"

    serving_size, servings = None, None

    try:
        res = requests.get(json_url, headers=headers, timeout=15)
        if res.status_code == 200:
            data = res.json()
            nutrition = data.get('recipe', {}).get('nutrition', {})
            serving_size = nutrition.get('servingSize')
            servings = nutrition.get('servingPerRecipe')
    except Exception:
        pass

    if serving_size is None and servings is None:
        try:
            res_html = requests.get(base_url, headers=headers, allow_redirects=True, timeout=15)
            canonical_url = res_html.url.rstrip('/')
            if canonical_url != base_url and "/recipe/" in canonical_url:
                canonical_json_url = f"{canonical_url}/as-json"
                res = requests.get(canonical_json_url, headers=headers, timeout=15)
                if res.status_code == 200:
                    data = res.json()
                    nutrition = data.get('recipe', {}).get('nutrition', {})
                    serving_size = nutrition.get('servingSize')
                    servings = nutrition.get('servingPerRecipe')
        except Exception:
            pass

    return index, serving_size, servings

# 2. Re-intento usando Multithreading
with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
    futures = {executor.submit(process_failed_recipe, idx): idx for idx in indices_pendientes}

    for future in tqdm(concurrent.futures.as_completed(futures), total=len(indices_pendientes)):
        idx, serving_size, servings = future.result()
        recipes.at[idx, 'ExtractedServingSize'] = serving_size if serving_size else "NOT_FOUND"
        recipes.at[idx, 'ExtractedServings'] = servings if servings else "NOT_FOUND"

# Verificamos el estado actual
pendientes_finales = (recipes['ExtractedServingSize'] == "NOT_FOUND").sum()
print(f"\nRecetas que definitivamente no tienen datos o fueron borradas: {pendientes_finales}")

# 3. Guardar el archivo definitivo y descargar
archivo_final = 'recipes_final_completo.csv'
recipes.to_csv(archivo_final, index=False)
print(f"\nDataset consolidado guardado como '{archivo_final}'. Iniciando descarga...")

# Forzar la descarga del archivo
files.download(archivo_final)
```

## Unificacion de RecipeServings y ExtractedServings

Sucedia que en el dataset original (https://www.kaggle.com/datasets/irkaal/foodcom-recipes-and-reviews?resource=download), tambien ocurria que varias recetas no contaban con "RecipeServings" por lo que ahora sí tenemos dicho dato para todas las recetas, que tambien se extrajo y se llama ExtractedServings.
Por lo que es necesario unificar todo, en una sola columna

```python
import numpy as np

# Veamos cuántos valores faltantes hay en cada columna originalmente
print(f"Faltantes en RecipeServings originales: {recipes['RecipeServings'].isna().sum()}")
print(f"Faltantes o NOT_FOUND en ExtractedServings: {(recipes['ExtractedServings'] == 'NOT_FOUND').sum() + recipes['ExtractedServings'].isna().sum()}")

# Reemplazamos los 'NOT_FOUND' por verdaderos nulos (NaN) para poder combinarlos fácilmente y evitamos el FutureWarning con infer_objects
recipes['ExtractedServings_clean'] = pd.to_numeric(recipes['ExtractedServings'].replace('NOT_FOUND', np.nan).infer_objects(copy=False), errors='coerce')

# Combinamos: si ExtractedServings_clean es nulo, usa el valor de RecipeServings
recipes['ConsolidatedServings'] = recipes['ExtractedServings_clean'].fillna(recipes['RecipeServings'])

# Vemos el resultado final
print(f"Faltantes en ConsolidatedServings final: {recipes['ConsolidatedServings'].isna().sum()}\n")

# Mostramos una muestra comparativa
display(recipes[['Name', 'RecipeServings', 'ExtractedServings', 'ConsolidatedServings']].head(15))
```
